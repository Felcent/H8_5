# Visualisasi

Selain menampilkan data dengan cara diatas, kita dapat memvisualisasikannya agar mudah dilihat. Cara visualisasinya pun beragam bisa dengan area plot, bar plot, pie plot, dsb. Berikut adalah salah satu contohnya.

```python
# Mengurutkan borough berdasarkan total value dalam urutan menurun
borough_sort = borough.sort_values(ascending=False)

# Memilih 3 borough dengan total kejahatan tertinggi
top_3_boroughs = borough_sort.head(3)

# Membuat daftar warna yang beragam
colors = ['red', 'green', 'blue']

# Membuat visualisasi
plt.figure(figsize=(5, 3))
top_3_boroughs.plot(kind='bar', color=colors)
plt.xlabel('Borough')
plt.ylabel('Total Kejahatan')
plt.title('3 Borough dengan Tingkat Kejahatan Tertinggi')
plt.xticks(rotation=0)
plt.show() 
```

## Area Plot

Question: Bagaimana perbandingan tingkat kejahatan di borough Harrow dan Greenwich dari tahun 2008 hingga 2012?

```python
# Filter data untuk tahun 2008-2012 dan borough Harrow, Greenwich dan Ealing
area_df = dataset[(dataset['year'] >= 2008) & (dataset['year'] <= 2012) & ((dataset['borough'] == 'Harrow') | (dataset['borough'] == 'Greenwich') | (dataset['borough'] == 'Ealing'))]

# Mengelompokkan data berdasarkan tahun, borough, dan menghitung total value
area_plt = area_df.groupby(['year', 'borough'])['value'].sum().unstack()

# Membuat visualisasi dengan area plot
plt.figure(figsize=(10, 4))
plt.fill_between(area_plt.index, area_plt['Harrow'], label='Harrow', alpha=0.5)
plt.fill_between(area_plt.index, area_plt['Greenwich'], label='Greenwich', alpha=0.5)
plt.fill_between(area_plt.index, area_plt['Ealing'], label='Ealing', alpha=0.5)

plt.xlabel('Tahun')
plt.ylabel('Total Kejahatan')
plt.title('Perbandingan Tingkat Kejahatan di Harrow, Greenwich dan Ealing untuk Tahun 2008-2012')

plt.legend(bbox_to_anchor = (1.2, 0.5))
plt.grid(True)

# Menampilkan grafik
plt.show()
```

Kesimpulan: Dapat dilihat bahwa borough Ealing adalah wilayah dengan total kejahatan terbanyak tiap tahunnya di antara ketiga borough diatas yang puncaknya berada di tahun 2010. Sedangkan untuk 2 borough lainnya memiliki tingkat kejahatan yang cukup stabil (Tidak terlalu naik/turun) dari tahun ke tahun.

## Histogram Plot

Question: Bagaimana perbandingan jenis kejahatan di tiap borough pada tahun 2010?

```python
# Filter data untuk tahun 2010 saja
df_2010 = dataset[dataset['year'] == 2010]

# Mengelompokkan data berdasarkan borough dan menghitung total value (untuk tahun 2010)
hist_plt = df_2010.groupby('major_category')['value'].sum()

# Membuat histogram plot
plt.figure(figsize=(8, 4))
plt.bar(hist_plt.index, hist_plt.values)
plt.xlabel('Major_category')
plt.ylabel('Total Kejahatan')
plt.title('Tingkat Kejahatan di Tiap Borough (2010)')
plt.xticks(rotation=65)  # Untuk memutar label borough agar lebih mudah dibaca
plt.grid(True)

# Menampilkan grafik
plt.show()
```

Kesimpulan: Dapat dilihat bahwa di antara semua major_category yang ada, Theft and Handling adalah jenis kejahatan yang terbanyak dilakukan pada tahun 2010.

## Bar Plot

Question: Bagaimana distribusi tingkat kejahatan major_category 'Robbery' dari tahun ke tahun?

```python
# Filter data untuk kategori "Robbery"
robbery_df = dataset[dataset['major_category'] == 'Robbery']

# Mengelompokkan data berdasarkan tahun dan menghitung total value
bar_plt = robbery_df.groupby('year')['value'].sum()

# Membuat bar chart vertical
plt.figure(figsize=(10, 6))
plt.barh(bar_plt.index, bar_plt.values, color='skyblue')
plt.xlabel('Total Kejahatan Robbery')
plt.ylabel('Tahun')
plt.title('Tingkat Kejahatan Robbery dari Tahun ke Tahun')
plt.grid(axis='y')

# Menampilkan grafik
plt.show()
```

Kesimpulan: Dapat dilihat bahwa jenis kejahatan Robbery cenderung mengalami peningkatan dari tahun 2008 hingga puncaknya terbanyak dilakukan pada tahun 2011 kemudian mulai mengalami penurunan sejak tahun 2012.

## Pie

Question: Secara keseluruhan, bagaimana distribusi presentase dari setiap jenis kejahatan yang dilakukan?

```python
# Membuat pie chart
plt.figure(figsize=(8, 7))
plt.pie(major, labels=major.index, autopct='%1.1f%%', startangle=90, pctdistance=0.80) # Variabel Major diambil dari atas
plt.axis('equal')  # Agar pie chart menjadi lingkaran
plt.title('Presentase Major Category Kejahatan')
plt.legend(bbox_to_anchor = (1, 0.5))

plt.show()
```

## Box Plot

Question: Secara keseluruhan, bagaimana distribusi dari setiap jenis kejahatan yang dilakukan?

```python
# Membuat box plot
plt.figure(figsize=(10, 6))
plt.boxplot([fuse_data[fuse_data['major_category'] == category]['value'] for category in fuse_data['major_category'].unique()], labels=fuse_data['major_category'].unique())

plt.xlabel('Major Category')
plt.ylabel('Jumlah Kejahatan')
plt.title('Distribusi Jenis Kejahatan')
plt.ylim(0, 30000)
plt.grid(True)

# Menampilkan grafik
plt.tight_layout()
plt.xticks(rotation=45)  # Agar label di sumbu x tidak tumpang tindih
plt.show()
```

Kesimpulan: Dapat dilihat bahwa Theft and Handling memiliki nilai maksimum tertinggi yang artinya Theft and Handling merupakan jenis kejahatan yang terbanyak dilakukan.

## Scatter Plot

Question: Apa tren kejahatan yang terjadi dari tahun ke tahun di London?

```python
# Mengelompokkan data berdasarkan major_category & year dan menghitung total value
total_df = dataset.groupby(['major_category', 'year'])['value'].sum().reset_index()

# Membuat scatter plot
plt.figure(figsize=(10, 6))
unique_categories = total_df['major_category'].unique()
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_categories)))
for i, major_category in enumerate(unique_categories):
    category_data = total_df[total_df['major_category'] == major_category]
    plt.scatter(category_data['year'], category_data['value'], label=major_category, color=colors[i], alpha=0.7)

plt.xlabel('Tahun')
plt.ylabel('Total Kejahatan')
plt.title('Tren Kejahatan dari Tahun ke Tahun')
plt.legend(bbox_to_anchor = (1, 0.5))
plt.grid(True)

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

Kesimpulan: Dapat dilihat bahwa tren kejahatan dari tahun ke tahun yang paling banyak dilakukan adalah Theft and Handling dan Violence Against the Person karena cenderung terus naik dari tahun ke tahun. Sedangkan Sexual Offences dan jenis kejahatan lainnya dapat dibilang cukup stabil tanpa peningkatan/penurunan yang cukup drastis.

## Word Clouds

Question: Bagaimana distribusi kejahatan berdasarkan minor_category dari tahun ke tahun

```python
# Mengelompokkan data berdasarkan major_category dan menghitung total value
minor_df = dataset.groupby('minor_category')['value'].sum().reset_index()

# Membuat word cloud
wordcloud_plt = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(minor_df.set_index('minor_category')['value'])

# Menampilkan word cloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud_plt, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud Jenis Kejahatan Berdasarkan Minor Category Terbanyak')
plt.show()
```

Kesimpulan: Dapat dilihat bahwa jenis kejahatan Other Theft dan Theft From Motor Vehicle adalah yang terbanyak dilakukan.
